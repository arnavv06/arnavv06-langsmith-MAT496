### Create a Dataset

Let's create a toy example dataset to run experiments over for a prompt

```python
from langsmith import Client

# A list of 5 Marvel-related questions and their answers
example_inputs = [
    ("Who is Iron Man?", "Tony Stark"),
    ("What is Captain America's shield made of?", "Vibranium"),
    ("Which infinity stone does Steven Strange guard?", "The Time Stone"),
    ("What is the name of Thor's hammer?", "Mjolnir"),
    ("Who wanted to destroy half the universe's population", "Thanos"),
]

client = Client()

dataset_name = "Marvel dataset"

dataset = client.create_dataset(
    dataset_name=dataset_name, description="Sample questions about the Marvel universe",
)


inputs = [{"question": input_prompt} for input_prompt, _ in example_inputs]
outputs = [{"output": output_answer} for _, output_answer in example_inputs]

client.create_examples(
    inputs=inputs,
    outputs=outputs,
    dataset_id=dataset.id,
)
```
